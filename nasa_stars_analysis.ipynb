```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_excel("newstar_planet12.xlsx")
df.shape
```

```python
print("total data count:",len(df))
print("\n my data columns names:")
print(df.columns.tolist())
```

```python
avg_temp = df['star_temperature_kelvi n'].mean()
print("stars average temperature(average temperature):",avg_temp)
```

```python
plt.figure(figsize=(8,5))
plt.hist(df['star_temperature_kelvi n'],bins=20,color='skyblue',edgecolor='black')
plt.title('distribution of star temperatures',fontsize=14)
plt.show
```

```python
import pandas as pd
import matplotlib.pyplot as plt
print("--- Kepler Magnitude Summary Statistics ---")
print(df['kepler_magnitude_brightness'].describe())
print("\nPotential dead or evolving stars (Highest Kepler Magnitude):")
dead_stars = df.sort_values(by='kepler_magnitude_brightness', ascending=False).head(5)
print(dead_stars[['kepler_magnitude_brightness', 'sky_right_ascension', 'sky_declination']])
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(9,6))
scatter = plt.scatter(df['sky_right_ascension'], df['sky_declination'], 
                      c=df['kepler_magnitude_brightness'], cmap='viridis_r', 
                      alpha=0.8, edgecolors='none', s=60)
cbar = plt.colorbar(scatter)
cbar.set_label('Kepler Magnitude Brightness (Higher = Dimmer)', fontsize=11)
plt.title('Spatial Distribution of Stars Across Celestial Coordinates', fontsize=13, pad=15)
plt.xlabel('Sky Right Ascension (RA)', fontsize=11)
plt.ylabel('Sky Declination (Dec)', fontsize=11)
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show
```
